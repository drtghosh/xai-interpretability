```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from dtreeviz.trees import *
import cufflinks as cf
```

```python
cf.set_config_file(sharing='public', theme='pearl', offline=False)
cf.go_offline()
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
```

```python
dt = DecisionTreeClassifier(min_samples_split=6)
dt.fit(X,y)
```

```python
print(f'Accuracy is: {dt.score(X,y)}')
```

```python
imp = pd.DataFrame({'feature': iris.feature_names, 'gini_importance': dt.feature_importances_})
imp = imp.sort_values(by=['gini_importance'], ascending=False)
imp.iplot(kind='bar', y='gini_importance', x='feature', color='purple', title='Feature Importance', xTitle='Feature names', yTitle='Gini importance')
```

```python
viz = dtreeviz(dt, X, y, feature_names=iris.feature_names, target_name='Species', class_names=iris.target_names.tolist())
viz
```

```python
X_useful = X[:,2:]
dt2 = DecisionTreeClassifier(min_samples_split=4)
dt2.fit(X_useful,y)
print(f'Accuracy is: {dt2.score(X_useful,y)}')
```

```python
viz = dtreeviz(dt2, X_useful, y, feature_names=iris.feature_names[2:], target_name='Species', class_names=iris.target_names.tolist())
viz
```
